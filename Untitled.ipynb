```python
import networkx as nx
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def load_graph_from_gml(file_path):
    # Load the graph from a GML file
    G = nx.read_gml(file_path)
    return G

def apply_kmeans(G, num_clusters):
    # Create a numpy array of the adjacency matrix
    A = nx.to_numpy_array(G)
    # Apply K-means clustering to the adjacency matrix
    kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(A)
    return kmeans.labels_

def group_nodes_by_cluster(G, labels):
    # Group nodes by their cluster labels
    clusters = {}
    for node, label in zip(G.nodes, labels):
        if label not in clusters:
            clusters[label] = []
        clusters[label].append(node)
    return clusters

def plot_graph(G, labels):
    # Plot the graph with node colors according to the cluster labels
    pos = nx.spring_layout(G)
    nx.draw_networkx_nodes(G, pos, node_color=labels, node_size=50, cmap=plt.cm.viridis)
    nx.draw_networkx_edges(G, pos, alpha=0.5)
    plt.show()

# Example usage:
file_path = 'dolphins.gml'  # Update this path to your actual file location
num_clusters = 30  # Adjust the number of clusters as needed
G = load_graph_from_gml(file_path)
labels = apply_kmeans(G, num_clusters)
clusters = group_nodes_by_cluster(G, labels)
plot_graph(G, labels)

# Print members of each cluster
for cluster_id, members in clusters.items():
    print(f"Cluster {cluster_id}: {members}")
```
